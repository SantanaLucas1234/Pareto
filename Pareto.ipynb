```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Criação de um dataset com duas coluna Tipo de serviço e faturamento
df = pd.DataFrame({"Tipo de Serviço" : ["Lojas online", "Serviços terceirizados",
                                        "Amazon Web Services", "Serviços de publicidade",
                                        "Serviços de assinatura", "Lojas Físicas",
                                        "Outro"],
                    "Faturamento - USD Bi" : [220, 117.72, 80.10, 37.74,
                                         35.22,18.96, 4.25]})

df
```

```python
df["Faturamento - USD Bi"].cumsum()
```

```python
#Criação de uma nova coluna - cumulativo
df["cum (%)"] = df["Faturamento - USD Bi"].cumsum()/df["Faturamento - USD Bi"].sum()*100
df
```

```python
#Criação do gráfico
fig,ax1 = plt.subplots(figsize = (8,6))

ax1.set_title('Faturamento de 2022 ')

color1 = 'tomato'
ax1.set_ylabel('Lucro',color = color1)

ax1.bar(df['Tipo de Serviço'], df['Faturamento - USD Bi'],color = color1,edgecolor = 'orange',linewidth = 2,\
       hatch = '*')
ax1.tick_params(axis = 'y',labelcolor = color1)

color2 = 'black'
ax2 = ax1.twinx()
ax2.set_ylabel('% Porcentagem',color = color2)

ax2.plot(df['Tipo de Serviço'], df['cum (%)'],color = color2,marker = 's',markersize = 8, linestyle = '-')

ax2.tick_params(axis = 'y',labelcolor = color2)
ax2.set_ylim([0,110])

for tick in ax1.get_xticklabels():
    tick.set_rotation(45)

plt.savefig('GraficoPareto2.png',format='png',dpi = 600, bbox_inches = 'tight')
plt.show()
```
